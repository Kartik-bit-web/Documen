```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/List_of_Indian_dishes#North-East_India'
r = requests.get(url).text

text = []
soup = BeautifulSoup(r, 'html.parser')
for table in soup.find_all('table'):
    for tbl in table.get('class'):
        if tbl == "sortable":
            add_it = []
            for i in table.find_all('tr'):   
                more_it = []
                for j in i.find_all('td'):
                    more_it.append(j)
                add_it.append(more_it)

            text.append(add_it)

print(text)
df = pd.DataFrame({})
df['details'] = np.array(text[0][1])





```
